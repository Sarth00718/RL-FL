```python
import random


value = [
    [0, 0, 0, 0],
    [0, 0, 0, 0],    #  0= Open Space, 1= Wall, 2= Goal, 3 = Door, 4= Key
    [0, 0, 0, 0],
    [0, 0, 0, 0],
        ]
```

```python
player_pos = [3, 0]
goal_pos = [0, 3]
```

```python
def next_state(current_state, action):
    if action=="up":
        if 0<=current_state[0]-1<=3 and 0<=current_state[1]<=3:
            return [current_state[0]-1, current_state[1]]
    elif action=="down":
        if 0<=current_state[0]+1<=3 and 0<=current_state[1]<=3:
            return [current_state[0]+1, current_state[1]]
    elif action=="left":
        if 0<=current_state[0]<=3 and 0<=current_state[1]-1<=3:
            return [current_state[0], current_state[1]-1]
    else:  # right
        if 0<=current_state[0]<=3 and 0<=current_state[1]+1<=3:
            return [current_state[0], current_state[1]+1]
    return current_state
```

```python
immediate_reward = -1
gamma = 0.9
pi = 0.25
actions = ["up", "down", "left", "right"]
```

```python
def policy_iteration(current_state, value):
    v = 0
    for action in actions:
        new_state = next_state(current_state, action)
        reward = 100 if new_state==goal_pos else immediate_reward
        v += pi*(reward + gamma*value[new_state[0]][new_state[1]])
    return v
```

```python
def print_matrix(mat):
    for i in range(4):
        print(mat[i])
```

```python
def fill_value_matrix(max_iterations):
    for m in range(max_iterations):
        for i in range(4):
            for j in range(3, -1, -1):
                if [i, j] == goal_pos:
                    continue
                value[i][j] = round(policy_iteration([i, j], value), 2)
        print_matrix(value)
        print()

```

```python
fill_value_matrix(3)
```
